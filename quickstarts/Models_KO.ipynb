## API 키 설정

다음 셀을 실행하려면 API 키가 `GEMINI_API_KEY`라는 이름의 Colab 시크릿에 저장되어 있어야 합니다. API 키가 없거나 Colab 시크릿 생성 방법을 모르는 경우, [인증](https://github.com/google-gemini/cookbook/blob/main/quickstarts/Authentication.ipynb) 예제를 참고하세요.

```
from google.colab import userdata

GEMINI_API_KEY = userdata.get("GEMINI_API_KEY")
```

## 모델 목록 조회

`list_models()`를 사용하여 사용 가능한 모델을 확인합니다. 이 모델들은 프롬프트에 사용되는 주요 메서드인 `generateContent`를 지원합니다.

```
from google import genai

client = genai.Client(api_key=GEMINI_API_KEY)

for model in client.models.list():
    print(model.name)
```

다음 모델들은 임베딩에 사용되는 `embedContent`를 지원합니다:

```
for model in client.models.list():
    if "embedContent" in model.supported_actions:
        print(model.name)
```

## 모델 세부 정보 확인

다음과 같이 `input_token_limit` 및 `output_token_limit`을 포함한 모델의 세부 정보를 확인할 수 있습니다.

```
for model in client.models.list():
    if model.name == "models/gemini-2.5-flash":
        print(model)
```
