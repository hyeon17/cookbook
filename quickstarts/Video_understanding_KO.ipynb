### 동영상 업로드

File API를 사용하여 모든 동영상을 업로드합니다. 사용 방법에 대한 자세한 내용은 [시작하기](../quickstarts/Get_started.ipynb#scrollTo=KdUjkIQP-G_i) 노트북을 참고하세요.

동영상을 처리하고 토큰화해야 하므로 몇 분 정도 걸릴 수 있습니다.

```
import time

def upload_video(video_file_name):
  video_file = client.files.upload(file=video_file_name)

  while video_file.state == "PROCESSING":
      print('Waiting for video to be processed.')
      time.sleep(10)
      video_file = client.files.get(name=video_file.name)

  if video_file.state == "FAILED":
    raise ValueError(video_file.state)
  print(f'Video processing complete: ' + video_file.uri)

  return video_file

pottery_video = upload_video('Pottery.mp4')
trailcam_video = upload_video('Trailcam.mp4')
post_its_video = upload_video('Post_its.mp4')
user_study_video = upload_video('User_study.mp4')
```

# YouTube 동영상 분석

직접 업로드한 동영상 외에도 Gemini에게 YouTube에서 동영상을 가져와 분석하도록 요청할 수 있습니다. 2023년 Google IO 기조연설을 예시로 사용해 보겠습니다. 주요 주제가 무엇이었는지 맞춰볼 수 있을까요?


```
response = client.models.generate_content(
    model=MODEL_ID,
    contents=types.Content(
        parts=[
            types.Part(text="Find all the instances where Sundar says \"AI\". Provide timestamps and broader context for each instance."),
            types.Part(
                file_data=types.FileData(file_uri='https://www.youtube.com/watch?v=ixRanV-rdAQ')
            )
        ]
    )
)

Markdown(response.text)
```

# 동영상 전처리 커스터마이징

Gemini API를 사용하면 동영상에서 정보를 이해하고 추출하는 능력을 향상시키기 위한 전처리 단계를 정의할 수 있습니다.

클리핑 인터벌(또는 특정 동영상 부분에 집중하기 위한 시간 오프셋 정의)과 커스텀 FPS(동영상 분석에 고려할 프레임 수 정의)를 사용할 수 있습니다.

이러한 기능에 대한 자세한 내용은 Gemini API 문서의 [동영상 전처리 커스터마이징](https://ai.google.dev/gemini-api/docs/video-understanding#customize-video-preprocessing)을 참고하세요.

## 클리핑 인터벌을 사용하여 동영상의 특정 부분 분석

때로는 동영상의 특정 부분을 찾고 싶을 수 있습니다. 요청에 시간 오프셋을 정의하여 모델에게 관심 있는 특정 동영상 구간을 알려줄 수 있습니다.

**참고:** 지정하는 `video_metadata`는 초 단위로 시간 오프셋을 나타내야 합니다.

이 예시에서는 [Google I/O 2025 기조연설](https://www.youtube.com/watch?v=XEzRZ35urlk) 동영상을 사용하고 모델에게 20분 50초에서 26분 10초 사이의 특정 시간 오프셋을 고려하도록 요청합니다.

```
response = client.models.generate_content(
    model=MODEL_ID,
    contents=types.Content(
        parts=[
            types.Part(
                file_data=types.FileData(file_uri='https://www.youtube.com/watch?v=XEzRZ35urlk'),
                video_metadata=types.VideoMetadata(
                    start_offset='1250s',
                    end_offset='1570s'
                )
            ),
            types.Part(text='Please summarize the video in 3 sentences.')
        ]
    )
)

Markdown(response.text)
```

클리핑 인터벌은 File API에 업로드된 동영상과 프롬프트에 인라인으로 포함된 동영상(인라인 데이터는 100MB를 초과할 수 없음을 기억하세요)에도 사용할 수 있습니다.

```
prompt = "Summarize this video in few short bullets"  # @param ["For each scene in this video, generate captions that describe the scene along with any spoken text placed in quotation marks. Place each caption into an object with the timecode of the caption in the video.", "Organize all scenes from this video in a table, along with timecode, a short description, a list of objects visible in the scene (with representative emojis) and an estimation of the level of excitement on a scale of 1 to 10"] {"allow-input":true}

video = trailcam_video # @param ["trailcam_video", "pottery_video", "post_its_video", "user_study_video"] {"type":"raw","allow-input":true}

response = client.models.generate_content(
    model=MODEL_ID,
    contents=types.Content(
        parts=[
            types.Part(
                file_data=types.FileData(
                    file_uri=video.uri,
                    mimeType=video.mime_type),
                video_metadata=types.VideoMetadata(
                    start_offset='60s',
                    end_offset='120s'
                )
            ),
            types.Part(text=prompt)
        ]
    )
)

Markdown(response.text)
```

## 초당 분석 동영상 프레임 수(FPS) 커스터마이징

기본적으로 Gemini API는 동영상 분석을 위해 1 FPS를 추출합니다. 하지만 이 양은 너무 많을 수 있습니다(강의처럼 활동이 적은 동영상의 경우) 또는 빠르게 변하는 시각적 내용에서 더 많은 세부 사항을 보존하려면 더 높은 FPS를 선택해야 합니다.

이 시나리오에서는 NASCAR 피트 스톱의 특정 구간을 사용하고 더 높은 FPS(이 경우 24 FPS)로 캡처합니다.

```
response = client.models.generate_content(
    model=MODEL_ID,
    contents=types.Content(
        parts=[
            types.Part(
                file_data=types.FileData(file_uri='https://www.youtube.com/watch?v=McN0-DpyHzE'),
                video_metadata=types.VideoMetadata(
                    start_offset='15s',
                    end_offset='35s',
                    fps=24
                )
            ),
            types.Part(text='How many tires where changed? Front tires or rear tires?')
        ]
    )
)

Markdown(response.text)
```
