* `examples/`의 콘텐츠에 대해 GitHub, 소셜 또는 선택한 사이트로의 링크 하나를 포함한 저자 정보를 선택적으로 추가할 수 있습니다.

### 코드 스타일

* 노트북은 사람을 위한 것입니다. 명확성을 위해 최적화된 코드를 작성하세요.
* 예시는 빠르고 간결하게 유지하세요.
* [Google Python 스타일 가이드](http://google.github.io/styleguide/pyguide.html)를 적용 가능한 곳에 사용하세요. [`pyink`](https://github.com/google/pyink)로 포맷된 코드는 항상 허용됩니다.
* 특히 함수와 변수를 정의할 때는 `=` 기호 주위에 추가 공백을 사용하지만, 함수 파라미터에는 사용하지 않습니다:
```python
    var = value
    function(
      parameter=value
    )
```
* 함수를 정의한 경우, 다른 함수에서 사용하기 전에 실행하고 결과를 보여주세요.
* 복잡한 것으로 결합하기 전에 작은 부분을 먼저 시연하세요.
* 코드를 가능한 한 단순하게 유지하고, 온도와 같은 추가 파라미터는 필요한 경우에만 사용하며 그 이유를 설명하세요
* 노트북 텍스트의 정확성을 보장하기 위해, 코드를 실행하여 모델 메타데이터를 표시하세요.
  * 예를 들어, 텍스트에 "1M 토큰 컨텍스트"라고 말하는 대신 `client.models.get(model='...').input_token_limit`의 출력을 표시하세요.

* 이 노트북에는 "이 노트북을 저장할 때 코드 셀 출력 생략" 옵션이 설정되어 있습니다. GitHub는 큰 diff(인라인 이미지)가 있는 노트북의 diff를 거부합니다.
* [ReviewNB.com](http://reviewnb.com)은 diff에 도움이 될 수 있습니다. 이는 노트북 풀 리퀘스트의 댓글에 링크됩니다.

* GitHub에서 노트북을 편집하는 가장 쉬운 방법은 편집하려는 브랜치에서 Colab으로 열고, 파일 --> GitHub에 사본 저장을 사용하면 열었던 브랜치에 저장됩니다.
